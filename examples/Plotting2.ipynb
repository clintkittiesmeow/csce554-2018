```julia
# Pkg.add("Distributions")
# Pkg.add("PyPlot")
using Distributions
using PyPlot
```

```julia
srand(1)
```

```julia
X() = sqrt(-2*log(rand()))*cos(2*pi*rand())
xGrid = -4:0.01:4
```

```julia
plt[:hist]([X() for _ in 1:10^6], 50, normed = true, ec="k", label="MC estimate")
PyPlot.plot(xGrid, pdf.(Normal(), xGrid), "-r", label="PDF")
xlim(-4, 4)
xlabel(L"$x$")
ylabel(L"f(x)")
legend(loc="upper right")
```
