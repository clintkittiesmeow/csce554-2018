```julia
#Pkg.clone("git@github.com:JuliaPlots/StatPlots.jl.git")
using DataFrames, IndexedTables
using StatPlots
gr(size=(500,400))
```

```julia
df1 = DataFrame(a = 1:10, b = 10*rand(10), c = 10 * rand(10))
```

```julia
@df df1 plot(:a, [:b :c], colour = [:red :blue])
```

```julia
@df df1 scatter(:a, :b, markersize = 4 * log.(:c + 0.1))
```

```julia
t = table(1:10, rand(10), names = [:a, :b]) # IndexedTable
```

```julia
@df t scatter(2 * :b)
```
